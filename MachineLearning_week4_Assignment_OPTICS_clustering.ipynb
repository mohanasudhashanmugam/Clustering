```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import OPTICS
op=OPTICS(min_samples=15, xi=0.05, max_eps=np.inf)
y_pred=op.fit_predict(x)
y_pred
```

```python
supervised=dataset
supervised['cluster_group']=y_pred
supervised.to_csv('cluster_optics.csv',index=False)
supervised.columns[3:6]
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```

```python
reachability = op.reachability_[op.ordering_]
plt.figure(figsize=(10, 5))
plt.plot(reachability)
plt.title('Reachability Plot')
plt.xlabel('Points (ordered)')
plt.ylabel('Reachability Distance')
plt.show()
```
