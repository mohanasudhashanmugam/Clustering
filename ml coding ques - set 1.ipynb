```python
#1.
Question 1:A retail store wants to identify customers who make frequent purchases. Given the dataset below, write a Python program to:
Group customers by their IDs.


Calculate the total purchase amount per customer.


Identify the top 3 customers with the highest purchase amounts.


Dataset:
data = {'Customer_ID': [101, 102, 103, 101, 104, 102, 101, 105, 102, 103],
        'Purchase_Amount': [200, 150, 180, 220, 300, 200, 100, 400, 250, 300]}

Expected Output:
Total Purchases per Customer:
   Customer_ID  Purchase_Amount
0          101              520
1          102              600
2          103              480
3          104              300
4          105              400

Top 3 Frequent Customers:
   Customer_ID  Purchase_Amount
1          102              600
0          101              520
2          103              480
```

```python
##Answer

import pandas as pd

dataset = {'Customer_ID': [101, 102, 103, 101, 104, 102, 101, 105, 102, 103],
        'Purchase_Amount': [200, 150, 180, 220, 300, 200, 100, 400, 250, 300]}

df = pd.DataFrame(dataset)
df_sum=df.groupby('Customer_ID',as_index=False)['Purchase_Amount'].sum()
print("Total Purchases per Customer:\n",df_sum,'\n')



df_count=df.groupby('Customer_ID').size().sort_values(ascending=False).head(3).rename('Count')

merged_df = pd.merge(df_sum,df_count, on='Customer_ID', how='inner').drop(columns='Count').sort_values('Purchase_Amount',ascending=False)

print("Top 3 Frequent Customers:\n",merged_df)

```

```python
#2. Predicting House Prices with Linear Regression
A real estate company wants to predict house prices based on square footage. Write a Python program to:
Train a Linear Regression model.


Predict house prices for given test data.


Dataset:
data = {'Square_Feet': [1500, 2000, 2500, 3000, 3500],
        'Price': [300000, 400000, 500000, 600000, 700000]}

Test Data: [[1800], [2800]]
Expected Output:
Predicted Prices:
[360000. 560000.]

```

```python
##2## Answer 
import pandas as pd

data = {'Square_Feet': [1500, 2000, 2500, 3000, 3500],
        'Price': [300000, 400000, 500000, 600000, 700000]}

df = pd.DataFrame(data)
#df.columns
independent=df[['Square_Feet']]
dependent=df[['Price']]

#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30)

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(independent,dependent)

y_pred=regressor.predict([[1800], [2800]])
print("Predicted Prices:\n",y_pred.flatten())
```

```python
#3:
Identifying Frequent Labels in a Dataset
A company wants to identify the top 3 most common categories in a dataset. Given the dataset below, write a Python program to:
Group the data by Category.


Count the total occurrences of each category.


Identify the top 3 most frequent categories.


Dataset:
data = {'Category': ['A', 'B', 'C', 'A', 'D', 'B', 'A', 'E', 'B', 'C', 'C', 'A'],
        'Value': [10, 15, 20, 30, 25, 18, 22, 40, 35, 50, 45, 15]}

Expected Output:

Total Occurrences per Category:
  Category  Count
0        A      4
1        B      3
2        C      3
3        D      1
4        E      1

Top 3 Frequent Categories:
  Category  Count
0        A      4
1        B      3
2        C      3
```

```python
##3##Answer
import pandas as pd

data = {'Category': ['A', 'B', 'C', 'A', 'D', 'B', 'A', 'E', 'B', 'C', 'C', 'A'],
        'Value': [10, 15, 20, 30, 25, 18, 22, 40, 35, 50, 45, 15]}

df = pd.DataFrame(data)
df_count=df.groupby('Category').size().reset_index(name='Count')
print("Total Occurrences per Category:\n",df_count,"\n")

print("Top 3 Frequent Categories:\n",df_count.sort_values('Count',ascending=False).head(3))

```

```python
#4.
Predicting Missing Values Using Mean Imputation
A dataset contains missing values in the Age column. Write a Python program to:
Replace missing values with the mean of the column.


Display the updated DataFrame.


Dataset:
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, None, 35, None]}

Expected Output:

Original Data:
     Name   Age
0  Alice  25.0
1    Bob  30.0
2 Charlie   NaN
3  David  35.0
4    Eve   NaN

Data after Imputation:
     Name   Age
0  Alice  25.0
1    Bob  30.0
2 Charlie  30.0
3  David  35.0
4    Eve  30.0
```

```python
##4#Answer

import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, None, 35, None]}
df=pd.DataFrame(data)
print("Original Data:\n",df,"\n")
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("Data after Imputation:\n",df)
```

```python
# 5: 
Implementing a Simple Linear Regression Model
You are given a dataset with Experience (years) and Salary ($). Write a Python program to:
Train a Linear Regression model.


Predict the salary for an individual with 6 years of experience.


Dataset:
import pandas as pd

data = {'Experience': [1, 2, 3, 4, 5], 
        'Salary': [30000, 35000, 40000, 45000, 50000]}
df = pd.DataFrame(data)

Expected Output (Example):

Predicted Salary for 6 years of experience: $55000


data = {'word_count': [100, 150, 200, 120, 180, 220], 
        'is_spam': ['ham', 'spam', 'spam', 'ham', 'spam', 'spam']}
df = pd.DataFrame(data)

Expected Output (Example):
Prediction for email with 200 words: Spam

```

```python
##5##Answer

##Predict the salary for an individual with 6 years of experience.

import pandas as pd

data = {'Experience': [1, 2, 3, 4, 5], 
        'Salary': [30000, 35000, 40000, 45000, 50000]}
df = pd.DataFrame(data)

independent=df[['Experience']]
dependent=df[['Salary']]

#from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(independent,dependent)

y_pred=regressor.predict([[6]])
print("Predicted Salary for 6 years of experience: $",y_pred.item())

```

```python
##5Answer##

import pandas as pd
data = {'word_count': [100, 150, 200, 120, 180, 220], 
        'is_spam': ['ham', 'spam', 'spam', 'ham', 'spam', 'spam']}
df = pd.DataFrame(data)

independent=df[['word_count']]
dependent=df[['is_spam']]

dependent.value_counts()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(dependent)
y_encoded

from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(independent,y_encoded)

y_pred_preoutput=classifier.predict([[200]])
y_pred=le.inverse_transform(y_pred_preoutput)
print("Prediction for email with 200 words: ",y_pred.item())
```
