```python
#8.
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Creating a dataset with missing values
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

# Find NaNs and replace them with 0
X_train[np.isnan(X_train)] = 0 

model = LinearRegression()
model.fit(X_train, y_train)  

model.predict([[6,8]])
```

```python
#10.
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

X_train = [[1, 100], [2, 200], [3, 300]]
y_train = [0, 1, 0]

sc = StandardScaler()
sc_X_train=sc.fit_transform(X_train)

model = SVC()
model.fit(sc_X_train, y_train)

pre_input=sc.transform([[1, 150]])#reversing the preprocessed input
print(pre_input)
print(model.predict(pre_input))
```
